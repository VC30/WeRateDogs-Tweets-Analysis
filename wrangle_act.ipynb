# Create a copy of each data set

```python
df_enhanced_clean = df_enhanced.copy()
df_images_tweet_clean = df_images_tweet.copy()
df_json_clean = df_json.copy()
```

# CLEANING

- Define: 
adjust the upper case in text column to title case

- Code:

```python
df_enhanced_clean['text'] = df_enhanced_clean['text'].str.title()
```

- Test:

```python
df_enhanced_clean['text'].head()
```

- Define: drop all rows containing RT using the retweeted_status_id column

Code: 

```python
df_enhanced_clean=df_enhanced_clean.drop(df_enhanced_clean[df_enhanced_clean.retweeted_status_id.isnull()== False].index)
```

- Test:

```python
df_enhanced_clean.text.head()
```

- Define: adjust the source column for readability

- Code:

```python
iPhone = '<a href="http://twitter.com/download/iphone" rel="nofollow">Twitter for iPhone</a>'
Vine = '<a href="http://vine.co" rel="nofollow">Vine - Make a Scene</a>'
Web = '<a href="http://twitter.com" rel="nofollow">Twitter Web Client</a>'
Deck = '<a href="https://about.twitter.com/products/tweetdeck" rel="nofollow">TweetDeck</a>' 


df_enhanced_clean.source = df_enhanced_clean.source.replace(iPhone, 'Twitter for iPhone')
df_enhanced_clean.source = df_enhanced_clean.source.replace(Vine, 'Vine - Make a Scene')
df_enhanced_clean.source = df_enhanced_clean.source.replace(Web, 'Twitter Web Client')
df_enhanced_clean.source = df_enhanced_clean.source.replace(Deck, 'TweetDeck')
```

- Test: 

```python
df_enhanced_clean.source.sample(10)
```

- Define:remove redundant columns

- Code:

```python
df_enhanced_clean = df_enhanced_clean[df_enhanced_clean.retweeted_status_id.isnull()]
df_enhanced_clean = df_enhanced_clean[df_enhanced_clean.retweeted_status_user_id.isnull()]
df_enhanced_clean = df_enhanced_clean[df_enhanced_clean.retweeted_status_timestamp.isnull()]
df_enhanced_clean = df_enhanced_clean[df_enhanced_clean.in_reply_to_status_id.isnull()]
df_enhanced_clean = df_enhanced_clean[df_enhanced_clean.in_reply_to_user_id.isnull()]
```

```python
df_enhanced_clean = df_enhanced_clean.drop(['retweeted_status_id','retweeted_status_user_id','retweeted_status_timestamp','in_reply_to_status_id','in_reply_to_user_id'], axis=1)
```

- Test:

```python
df_enhanced_clean.columns
```

```python
df_enhanced_clean.head()
```

- Define: change timestamp format to datetime

- Code: 

```python
df_enhanced_clean['timestamp'] = df_enhanced_clean['timestamp'][0:-6]     #.str.slice(start=0, stop=-6)
```

```python
df_enhanced_clean['timestamp'] = pd.to_datetime(df_enhanced_clean['timestamp'], format = "%Y-%m-%d %H:%M:%S")
```

- Test

```python
df_enhanced_clean['timestamp'].dtype
```

- Define: remove duplicates in name column

- Code:
- Test:

```python
#df_enhanced_clean1 = df_enhanced_clean.copy()
df_enhanced_clean.name.drop_duplicates(keep=False)
```

Define: replace - by _ in p1,p2 and p3

Code:

```python
df_images_tweet_clean.p1.replace({'-':'_'}, inplace=True)
df_images_tweet_clean.p2.replace({'-':'_'},inplace=True)
df_images_tweet_clean.p3.replace({'-':'_'},inplace=True)
```

Test:

```python
df_images_tweet_clean.head()
```

Define: rename column from id to tweet_id

Code: 

```python
df_json_clean = df_json_clean.rename(columns = {'id':'tweet_id'})
```

Test: 

```python
df_json_clean.head()
```

Define: combine the doggo, floofer, pupper and puppo columns into a single column 'dog_type'

Code: 

```python
df_enhanced_clean['dog_type']=df_enhanced_clean.doggo.replace('None','')+df_enhanced_clean.floofer.replace('None','')+df_enhanced_clean.pupper.replace('None','')+df_enhanced_clean.puppo.replace('None','')
```

Test:

```python
df_enhanced_clean.dog_type
```

Define: Drop the doggo, floofer, pupper and puppo columns

Code: 

```python
df_enhanced_clean = df_enhanced_clean.drop(['doggo','floofer','pupper','puppo'],axis = 1)
```

Test: 

```python
df_enhanced_clean.columns
```

```python
Define: Merge the three datasets to form a master dataset
```

Code:

```python
df_compiled_clean = df_enhanced_clean.merge(df_images_tweet_clean,on='tweet_id').merge(df_json_clean,on='tweet_id')
```

# Analyzing, and Visualizing Data

```python
df = pd.read_csv('twitter_archive_master.csv')
df.head(2)
```

```python
#Top 10 dogs with the highest mean retweets
df.groupby('name')['retweet_count'].mean().sort_values(ascending=False).head(10)
```

```python
#Top 10 dogs with the highest mean retweets
df.groupby('name')['favorite_count'].mean().sort_values(ascending=False).head(10)
```

```python
#Top 10 dogs with the highest mean rating numerator
df.groupby('name')['rating_numerator'].mean().sort_values(ascending=False).head(10)
```

```python
#Top 10 dogs with the highest mean rating denominator
df.groupby('name')['rating_denominator'].mean().sort_values(ascending=False).head(10)
```

```python
import sklearn
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=df.retweet_count
y=df.favorite_count

plt.title('Correlation between Retweet Count and Favorite Count')
plt.scatter(x, y)
plt.plot(np.unique(x),np.poly1d(np.polyfit(x, y, 1))(np.unique(x)), color='red')
plt.xlabel('Retweet Count')
plt.ylabel('Favorite Count')
plt.show()
```
